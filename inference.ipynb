### Imports


```python
import json

import numpy as np
import torch
import librosa
from librosa import power_to_db

from IPython.display import Audio
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
from dataset import NSynthDataset
from models import Generator
from utils import AttrDict, mel_spectrogram
```

### Setup


```python
with open("config.json") as f:
        data = f.read()

json_config = json.loads(data)
cfg = AttrDict(json_config)

torch.manual_seed(cfg.train["seed"])
torch.cuda.manual_seed(cfg.train["seed"])
```

```python
device = torch.device("cuda:0") if torch.cuda.is_available() else torch.device("cpu")
```

```python
dataset = NSynthDataset(split="validation", shuffle=False, cfg=cfg)
```

### Utilies


```python
def get_pair(i, j):
  x, y = dataset[i], dataset[j]
  return x.unsqueeze(0).to(device, non_blocking=True), y.unsqueeze(0).to(device, non_blocking=True)

def play_audio(x):
  x = x.cpu().detach().numpy().astype(float).reshape(-1)
  display(Audio(x, rate=cfg.audio["sr"]))

def display_mel(content, style, generated):

  fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(20, 8))
  fig.dpi = 300

  content_mel = power_to_db(mel_spectrogram(content.squeeze(), cfg), ref=np.max)

  librosa.display.specshow(
            content_mel,
            sr=cfg.audio["sr"],
            x_axis="time",
            y_axis="mel",
            fmax=24000,
            ax=axes[0],
  )

  style_mel = power_to_db(mel_spectrogram(style.squeeze(), cfg), ref=np.max)

  librosa.display.specshow(
            style_mel,
            sr=cfg.audio["sr"],
            x_axis="time",
            y_axis="mel",
            fmax=24000,
            ax=axes[1],
  )

  generated_mel = power_to_db(mel_spectrogram(generated.squeeze(), cfg), ref=np.max)

  librosa.display.specshow(
            generated_mel,
            sr=cfg.audio["sr"],
            x_axis="time",
            y_axis="mel",
            fmax=24000,
            ax=axes[2],
  )

  fig.tight_layout()
  plt.show()
```

### Model Output


```python
gen = Generator(cfg).to(device)
checkpoint = "weights/gen_only"
state_dict = torch.load(checkpoint, map_location=device)
gen.load_state_dict(state_dict["gen"])
```

#### Example 1


```python
gen.eval()
torch.cuda.empty_cache()

with torch.no_grad():
  content, style = get_pair(10, 27)
  generated = gen(content, style, [1, 1, 1, 1, 0.2])
```

```python
for audio in (content, style, generated):
  play_audio(audio)
```

```python
display_mel(content, style, generated)
```
